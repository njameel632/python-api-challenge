```python
# Dependencies

import pandas as pd
import requests
import csv
import seaborn as sb
import numpy as np
!pip install citipy
from citipy import citipy
!pip install RateLimiter
from ratelimiter import RateLimiter
from config import api_key
import json
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Generating Random Latitudes & Longitudes

counter = 0
rndlat = []
rndlon = []

while len(rndlat) < 1500:
    lat = np.random.uniform(-90, 90)
    lon = np.random.uniform(-180, 180)
    rndlat.append(lat)
    rndlon.append(lon)
    counter += 1
    
geo_coordinates = pd.DataFrame({"lat": rndlat, "lon":rndlon})
geo_coordinates.head()
```

```python
# Feeding the Latitudes & Longitudes to Citipy

city = []
country = []

for index, row in geo_coordinates.iterrows():
    city_lat = (row["lat"])
    city_lon = (row["lon"])
    city_name = citipy.nearest_city(city_lat, city_lon).city_name
    country_code = citipy.nearest_city(city_lat, city_lon).country_code
    city.append(city_name)
    country.append(country_code)

city_data = pd.DataFrame({"city": city, "country": country})
city_data = city_data.drop_duplicates()
city_data.head()
```

```python
# We now have more than 500 cities

weather_data = []
len(city_data)
```

```python
# Making calls to APIs

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

query_url = f"{url}appid={api_key}&units={units}&q="
rate_limiter = RateLimiter(max_calls=50, period=60)


for index, row in city_data.iterrows():
    with rate_limiter:
        city_q=row["city"]
        response = requests.get(query_url + city_q).json()
        record_counter=1
        print (f"Processing record #{record_counter} -- {city_q}")
        print (f'{url}appid=***masked***&units={units}&q={city_q}')
        record_counter +=1
        
        try:
            city_lon=response["coord"]["lon"]
            temp=response["main"]["temp"]
            humid=response["main"]["humidity"]
            cloud=response["clouds"]["all"]
            wind=response["wind"]["speed"]
            city_lat=response["coord"]["lat"]
            city_name=response["name"]
            city_country=response["sys"]["country"]
            weather_data.append({"City": city_name, "Country": city_country, "Latitude": city_lat, "Longitude": city_lon, 
                                 "Temperature (F)": temp, "Humidity (%)": humid, "Cloudiness (%)":cloud, "Wind Speed (MPH)": wind})
            
        except:
            print (f"Weather for {city_q} not found.")
            print ("----------------------------------------------------------------")
            continue
```

```python
weather_data = pd.DataFrame(weather_data)
```
